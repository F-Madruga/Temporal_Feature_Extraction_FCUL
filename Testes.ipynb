```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./dataset.csv", names=["INDEX", "YEAR", "MONTH", "DAY", "AMOUNT", "BUYER", "SELLER", "CREDIT_CARD", "LATITUDE", "LONGITUDE"])
df
```

```python
seller = 302;
buyer = 18;
year = 2018;
month = 10;
```

```python
df[df["SELLER"] == seller].mean().to_frame().transpose()
```

```python
df[df["BUYER"] == buyer].mean().to_frame().transpose()
```

```python
# Filter by seller
seller_df = df[df["SELLER"] == seller]

# Then filter by month
month_seller_df = seller_df[seller_df["MONTH"] == month]

# Then filter by year
year_month_seller_df = month_seller_df[month_seller_df["YEAR"] == year]

# Then calculate mean
year_month_seller_df.mean().to_frame().transpose()
```
